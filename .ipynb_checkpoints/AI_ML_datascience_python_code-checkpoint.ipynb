```python
import numpy as np
l1=np.arange(1,11).reshape(2,5)
l2=np.arange(1,11).reshape(2,5)
print(l1)
print(l2)
print(l1+l2)
print(np.add(l1,l2))
print(l1-l2)
print(np.subtract(l1,l2))
print(l1/l2)
print(np.divide(l1,l2))
print(l1*l2)
print(np.multiply(l1,l2))
print(l1.max())
print(l2.max())
print(l1.argmax())
print(l2.argmax())
print(l1.max(axis=0))#column wise max values
print(l2.max(axis=0))
print(l1.max(axis=1))#row wise max values
print(l2.max(axis=1))
print(l1.min(axis=0))#column wise min values
print(l2.min(axis=0))
print(l1.min(axis=1))#row wise min values
print(l2.min(axis=1))
print(l1.min())
print(l2.min())
print(l1.argmin())
print(l2.argmin())
print(np.sum(l1))
print(np.sum(l2))
print(np.sum(l1,axis=0))
print(np.sum(l1,axis=1))
print(np.sum(l2,axis=0))
print(np.sum(l2,axis=1))
print(np.mean(l1))
print(np.mean(l2))
print(np.sqrt(l1))
print(np.sqrt(l2))
print(np.std(l1))
print(np.std(l2))
print(np.exp(l1))
print(np.exp(l2))
print(np.log(l1))
print(np.log(l2))
print(np.log10(l1))
print(np.log10(l2))
```
