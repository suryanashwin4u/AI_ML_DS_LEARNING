```python
# linear search using function
# no need to sort array
# worst case complexity bigO(n)

def linearsearch(mylist,length,key): 
    for i in range(0,length):
            if(mylist[i]==key):
                return i
    return -1  

mylist=input("enter numbers in an array:").split(",")
key=input("enter the key to search from the list:")
length=len(mylist)
index=linearsearch(mylist,length,key)
if(index==-1):
    print("{Sorry, key is not found in the list")
else:
    print(f"key:{key} is found at {index+1}th position")

 
```

```python
# binary search using recursion
# array should be sorted before doing binary search
# worst case complexity bigO(log (n))

def binarysearch(mylist,key,leftindex,rightindex):
   
    if leftindex<=rightindex:
        mid=(rightindex+leftindex)//2
        if mylist[mid]==key:
            return mid
        elif mylist[mid]<key:
            return binarysearch(mylist,key,mid+1,rightindex)
        elif mylist[mid]>key:
            return binarysearch(mylist,key,leftindex,mid-1)
        
    return -1
    
mylist=input("enter the numbers in a sorted way:").split(",")
print(f"Numbers entered in the list:{mylist}")
key=input("enter the key you want to search:")

leftindex=0
rightindex=len(mylist)-1

index=binarysearch(mylist,key,leftindex,rightindex)

if index==-1:
    print("sorry, the given key is not found in the list")
else:
    print(f"key is found at {index+1} position")
```

```python
# binary search using iteration 
# array should be sorted before doing binary search
# worst case complexity bigO(log (n))


def binarysearch(mylist,key,leftindex,rightindex):
   
    while leftindex<=rightindex:
        mid=(rightindex+leftindex)//2
        if mylist[mid]==key:
            return mid
        elif mylist[mid]<key:
            leftindex=mid+1
        elif mylist[mid]>key:
            rightindex=mid-1
        
    return -1
    
mylist=input("enter the numbers in a sorted way:").split(",")
print(f"Numbers entered in the list:{mylist}")
key=input("enter the key you want to search:")

leftindex=0
rightindex=len(mylist)-1

index=binarysearch(mylist,key,leftindex,rightindex)

if index==-1:
    print("sorry, the given key is not found in the list")
else:
    print(f"key is found at {index+1} position")

```

```python
# jump search
# array should be sorted for jump search
# n=size of list and m=block size
# no. of jumps=n/m and m-1=no. of linear searches done
# comparisons in the worst case (n/m+(m-1))
# comparisons will be minimum when m=root(n)
# optimal block size in the worst case root(n)
# time complexity=O(root(n))
# comparing time complexity: linear search<jump search<binary search


import math
def jumpsearch(mylist,key,length):
    step=math.sqrt(length)                            #finding block size to jump
    previous=0                                        
    while mylist[int(min(step,length)-1)]<key:        #finding block where element is present
        previous=step
        step=step+math.sqrt(length)
        if previous>=length:
            return -1
    while mylist[int(previous)]<key:                  #doing a linear search
        previous=previous+1
        if previous==min(step,length):
            return -1
    if mylist[int(previous)]==key:
        return previous
    return -1

mylist=input("enter the numbers in a sorted way:").split(",")
print(f"Numbers entered in the list:{mylist}")
key=input("enter the key you want to search:")
length=len(mylist)

index=jumpsearch(mylist,key,length)

if index==-1:
    print("sorry, the given key is not found in the list")
else:
    print(f"key is found at {index+1} position")
```

```python
# interpolation search using function
# uniformaly distributed and sorted array needed
# position formula-->position=lower_index+[{key-mylist(lower_index)}*{(higher_index-lower_index)}]/(mylist[higher_index]-mylist[lower_index])
# if uniformaly distributed then complexity is O(log (log (n)) and in worst case it is O(n)

def interpolation_search(mylist,list_length,key):
    lower_index=0
    higher_index=list_length-1
    while lower_index<=higher_index and key>=mylist[lower_index] and key<=mylist[higher_index]:
        if lower_index==higher_index:
            if mylist[lower_index]==key:
                return lower_index
            return -1
        
        position  = lower_index + int(((float(higher_index - lower_index) / ( mylist[higher_index] - mylist[lower_index])) * ( key - mylist[lower_index])))
        if mylist[position]==key:
            return position
        elif mylist[position]<key:
            lower_index=position+1
        elif mylist[position]>key:
            higher_index=position-1
    return -1



mylist=input("enter the sorted and uniformaly distributed list of numbers:").split(",")
print(f"list input:{mylist}")

key=int(input("enter the searching element:"))
print(f"key entered is:{key}")

list_length=len(mylist)
print(f"length of the list:{list_length}")

index=interpolation_search(mylist,list_length,key)

if index==-1:
    print("element not found")
else:
    print(f"element is found at {index+1}")
```

```python
# interpolation search using recursion
# uniformaly distributed and sorted array needed
# position formula-->position=lower_index+[{key-mylist(lower_index)}*{(higher_index-lower_index)}]/(mylist[higher_index]-mylist[lower_index])
# if uniformaly distributed then complexity is O(log (log (n)) and in worst case it is O(n)

def interpolation_search(mylist,lower_index,higher_index,key):
    lower_index=0
    higher_index=list_length-1
    if lower_index<=higher_index and key>=mylist[lower_index] and key<=mylist[higher_index]:
        position  = lower_index + int(((float(higher_index - lower_index) / ( mylist[higher_index] - mylist[lower_index])) * ( key - mylist[lower_index])))
        if mylist[position]==key:
            return position
        elif mylist[position]<key:
            return interpolation_search(mylist,position+1,higher_index,key)
        elif mylist[position]>key:
            return interpolation_search(mylist,lower_index,position-1,key)
    return -1

mylist=input("enter the sorted and uniformaly distributed list of numbers:").split(",")
print(f"list input:{mylist}")

key=input("enter the searching element:")
print(f"key entered is:{key}")

list_length=len(mylist)
print(f"length of the list:{list_length}")

index=interpolation_search(mylist,0,list_length,key)

if index==-1:
    print("element not found")
else:
    print(f"element is found at {index+1}")
```

```python
# exponential search
# time complexity is log(n)
# useful for unbounded(infinite amount) searches
# works better than binary search for bounded arrays and also when the element to be searched is closer to the first element.

def exponential_search(mylist,list_length,key):
    if mylist[0]==key:
        return 0
    i=1
    while i<list_length and mylist[i]<=key:
        i=i*2
    return binary_search(mylist,i/2,min(i,list_length),key)

def binary_search(mylist,left_index,right_index,key):
    if right_index>=left_index:
        mid=int(left_index+(right_index-left_index)/2)
        if mylist[mid]==key:
            return mid
        if mylist[mid]>key:
            return binary_search(mylist,left_index,mid-1,key)
        if mylist[mid]<key:
            return binary_search(mylist,mid+1,right_index,key)
    return -1


mylist=input("enter numbers to be sorted:").split(",")
list_length=len(mylist)
key=input("enter the key element:")
index=exponential_search(mylist,list_length,key)
if index==-1:
    print("element not found")
else:
    print(f"element is found at index:{index}")
```

```python
# recursive program to linearly search an element in a given array using recursion

def linear_search_recursive(mylist,left_index,right_index,key):
    if right_index<left_index:
        return -1
    if mylist[left_index]==key:
        return left_index
    if mylist[right_index]==key:
        return right_index
    return linear_search_recursive(mylist,left_index+1,right_index-1,key)

mylist=input("enter numbers in the list:").split(",")
left_index=0
right_index=len(mylist)-1
key=input("enter the key element:")
index=linear_search_recursive(mylist,left_index,right_index,key)
if index==-1:
    print("element not found")
else:
    print(f"element is found at index:{index+1}")
```
